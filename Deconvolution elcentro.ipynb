```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les fichiers .txt avec sep='\s+'
inputelcentro = pd.read_csv('correctedelcentroinput.txt', header=None, sep=r'\s+')
deconvolvedelcentro = pd.read_csv('deconvolvedelcentro.txt', header=None, sep=r'\s+')

# Supposons que la premire colonne soit le temps et la deuxime l'acclration
time1 = inputelcentro[0]
accel1 = inputelcentro[1]

time2 = deconvolvedelcentro[0]
accel2 = deconvolvedelcentro[1]

# Tracer les deux courbes sur le mme graphique
plt.figure(figsize=(10, 6))
plt.plot(time1, accel1, label='Corrected El Centro ground motion', color='orange')
plt.plot(time2, accel2, label='Deconvolved El Centro', color='grey')

# Ajouter des labels et un titre
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s)')
plt.title('Comparison of El Centro and Deconvolved El Centro Acceleration')
plt.legend()
plt.savefig("comparaison Elcentro.png")
# Afficher le graphique
plt.grid(True)
plt.show()

plt.show()
```
